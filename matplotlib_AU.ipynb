```
import numpy as np
import matplotlib.pyplot as plt

ironman=np.linspace(-10,10,100)
fig=plt.figure()

plt.plot(ironman, np.sin(ironman), '.')
fig=plt.figure()
plt.plot(ironman, np.cos(ironman), '-')
```
